```python
import numpy as np
import matplotlib.pyplot as plt
# Program written by Nicholas Munoz
y =[0.,10.]
t=0.0
tend=5
h = 1e-4
bool = 1
zp=[]
tp=[]
```

```python
def f(t, y):
    f=np.zeros(2)
    f[0]=y[1]
    f[1]= -9.8
    return f
    
def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t,y)
    k2=h*f(t+0.5*h,y+0.5*k1)
    k3=h*f(t+0.5*h,y+0.5*k2)
    k4=h*f(t+h,y+k3)
    y+=(k1+2*(k2+k3)+k4)/6
    return y
while(bool==1):
    if(y[0]>=0):
        y = rk4(t,h,y,4)
        zp=np.append(zp,y[0])
        tp=np.append(tp,t)
        t = t+h
        bool = 1
    else:
        bool = 0
plt.plot(tp,zp)
plt.title("Trajectory of Ball thrown upwards")
plt.xlabel("time")
plt.ylabel("position")
plt.show()
```
