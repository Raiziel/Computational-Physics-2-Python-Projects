```python
import matplotlib.pylab as plt
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy as np;
```

```python
omega= 0.9
mu = 100
Nmax = 100; Niter = 200 
V = np.zeros((Nmax, Nmax), float)  
V[50,50] = -mu     

x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)                 
```

```python
def functz(V):  
    z = V[X,Y]                        
    return z
```

```python

Z = functz(V)                          
fig = p.figure(figsize=[8,10])                       
ax = Axes3D(fig)                                     
ax.plot_wireframe(X, Y, Z, color = 'r')              
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()    


for iter in range(Niter):                                  
    if iter%10 == 0: print(iter)
    maxerr=0.
    for i in range(1, Nmax-1):                                                
        for j in range(1,Nmax-1): 
            du = 1/4*(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])-V[i,j]
            if(maxerr<np.abs(du)):
                maxerr=np.abs(du)
            V[i,j]=V[i,j]+omega*du
    
    if(maxerr<1.e-3):
        break
            


Z = functz(V)                          
fig = p.figure(figsize=[8,10])                        # Create figure
ax = Axes3D(fig)                                      # Plot axes
ax.plot_wireframe(X, Y, Z, color = 'r')               # Red wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()
```

```python
def ln(x, y, mu):
    r = sqrt((x-50.)*(x-50.)+(y-50.)*(y-50.))
    
    return -2*mu*np.log(r)+2.5*np.pi
```

```python
Zln = ln(X,Y,mu)
fig = p.figure()                                      # Create figure
ax = Axes3D(fig)                                      # Plot axes
ax.plot_wireframe(X, Y, Zln, color = 'b')              # Blue wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()                                              # Show fig
```
