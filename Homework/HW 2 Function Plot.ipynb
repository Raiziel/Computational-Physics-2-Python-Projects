```python
# Program Written by Nicholas Munoz
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
f = lambda x,N : (math.pi/2)+2*(np.sin(x)+(np.sin(3*x)/3)+(np.sin(5*x)/5)+ (np.sin(x*N)/N))
N=11
x = np.arange(-math.pi,math.pi,0.1)
plt.plot(x,f(x,N))
N=21
plt.plot(x,f(x,N))
N=31
plt.plot(x,f(x,N))
plt.title("Function Plot")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
x = np.arange(-math.pi,math.pi,0.1)
def FunctionPlot(x,N):
    f = lambda x,N : (math.pi/2)+2*(np.sin(x)+(np.sin(3*x)/3)+(np.sin(5*x)/5)+ (np.sin(x*N)/N))
    plt.plot(x,f(x,N))
    plt.title("Function Plot")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
```

```python
FunctionPlot(x,11)
FunctionPlot(x,21)
FunctionPlot(x,31)
```
