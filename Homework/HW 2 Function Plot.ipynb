```python
# Program Written by Nicholas Munoz
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
f = lambda x,N : (math.pi/2)+2*(np.sin(x)+(np.sin(3*x)/3)+(np.sin(5*x)/5)+ (np.sin(x*N)*(1./N)))

N=11
x = np.arange(-math.pi,math.pi,0.1)
for i in range (1,N+1): #evaluates from 1-11
    plt.plot(x,f(x,i)) 
N=21
for i in range (1,N+1): #evaluates from 1-21
    plt.plot(x,f(x,i))
N=31
for i in range (1,N+1): #evaluates from 1-31
    plt.plot(x,f(x,i))
plt.title("Function Plot")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
x = np.arange(-math.pi,math.pi,0.1)
def FunctionPlot(x,N):
    f = lambda x,N : (math.pi/2)+2*(np.sin(x)+(np.sin(3*x)/3)+(np.sin(5*x)/5)+ (np.sin(x*N)/N))
    for i in range (1,N+1):
        plt.plot(x,f(x,i))
        N = str(i)
        plt.title('Function Plot of N = %i '%i)
        plt.xlabel("x")
        plt.ylabel("f(x)")
        plt.show()
```

```python
FunctionPlot(x,11)
```

```python
FunctionPlot(x,21)
```

```python
FunctionPlot(x,31)
```
