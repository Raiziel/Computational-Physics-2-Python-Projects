<B> Write a program to numerically evaluate the derivative of f(x)=cos x at an arbitrary position x.
1) Evaluate the derivative at position x using the forward difference scheme. Plot log(Error)=log10|f’(numerically)-f’(analytic)| against log10(dx) for dx=1e-16,1e-15,….,0.1 at x=π/4. You should find Error~$dx^n$; what is n? what is the optimal value for dx.
    
   
2) Repeat the exercise using the central difference scheme. Why this scheme is better?


3) Repeat (2) for x=0. Why do the results differ? (hint: write out the error term derived in the class explicitly by plugging the derivative of cos or sin) <B>

```python
# Program Written by Nicholas Munoz
import numpy as np
import matplotlib.pyplot as plt
def analyticsol(x):
    analytic = -np.sin(x)
    return analytic
dx=1*10**(np.sort(-np.arange(1, 17.)))
```

```python
#forward
def deriv1(f,x,dx):
    return (f(x+dx) - f(x)) / dx
```

```python
#central
def deriv2(f,x,dx):
    return (f(x+dx)/2 - f(x-dx)/2)/dx
```

```python
error=[]
for i in dx:
    numerical = deriv1(f=np.cos, x=np.pi/4, dx=i)
    calc = np.log10(np.abs(numerical - analyticsol(x=np.pi/4)))
    error.append(calc)
    
plt.title('Error using Forward Method')
plt.xlabel('dx')
plt.ylabel('error')
plt.grid()
plt.plot(np.log10(dx),error)

ymin = min(error)
ymax = max(error)
xmin = np.log10(min(dx))
xmax = np.log10(max(dx))

n = ((ymax-ymin)/(xmax-xmin))
print ('n = ',2*n)

print('optimal dx is', dx[error.index(min(error))])
```

```python
error=[]
for i in dx:
    numerical = deriv2(f=np.cos, x=np.pi/4, dx=i)
    calc = np.log10(np.abs(numerical - analyticsol(x=np.pi/4)))
    error.append(calc)
    
plt.title('Error using Central Method')
plt.xlabel('dx')
plt.ylabel('error')
plt.grid()
plt.plot(np.log10(dx),error)

ymin = min(error)
ymax = max(error)
xmin = np.log10(min(dx))
xmax = np.log10(max(dx))

n = ((ymax-ymin)/(xmax-xmin))
print ('n = ',2*n)
print('optimal dx is', dx[error.index(min(error))])
```

<B> The central difference method is better because it is a higher order than the forward difference where central difference is h^2 while forward method is order h <B>

```python
error2=[]
for i in dx:
    numerical = deriv2(f=np.cos, x=0, dx=i)
    calc = np.log10(np.abs(numerical - analyticsol(x=0))) 
    error2.append(calc)
    
plt.title('Error using Central Method')
plt.xlabel('dx')
plt.ylabel('error')
plt.grid()
plt.plot(np.log10(dx),error2)



```
